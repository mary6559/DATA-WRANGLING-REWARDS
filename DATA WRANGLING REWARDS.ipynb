# I’m importing libraries and downloading data

```python
import pandas as pd

plik = r"C:\Users\maria\Downloads\RewardsData.csv"
df = pd.read_csv(plik)
df.head(20)
```

# I search for whether the records are NaN

```python
df.isna()
```

# I count how many missing values there are in each column

```python
missing_value_counts = df.isna().sum()
missing_value_counts
```

# I calculate the percentage of missing values

```python
missing_value_percentages = missing_value_counts / len(df) * 100
missing_value_percentages
```

# I check the column I want to clean (fill in missing values) to see its data type, as well as its minimum and maximum values

```python
df['Last Seen']
```

```python
# I change the data type to datetime
df['Last Seen'] = pd.to_datetime(df['Last Seen'])
df['Last Seen']
```

```python
df['Last Seen'].max()
```

```python
df['Last Seen'].min()
```

# I fill the missing values with the minimum

```python
df['Last Seen'].fillna(value = df['Last Seen'].min(), inplace=True)
```

```python
missing_value_counts = df.isna().sum()
missing_value_counts
```

# I remove all records that have NaN values

```python
missing_value_counts = sample_df.isna().sum()
missing_value_counts
```

# I remove unnecessary columns

```python
sample_df = sample_df.drop(columns = 'Tags')
sample_df
```

```python
city_unique_values = sample_df['City'].unique()
city_unique_values.sort()
city_unique_values
```

# I convert the us_state_to_abbrev dictionary into a DataFrame so that it can be merged with sample_df

```python
states_df = pd.DataFrame(list(us_state_to_abbrev.items()),columns  = ['State Name', 'State'])
states_df.head()
```

# I merge the original DataFrame with the states DataFrame

```python
sample_df = pd.merge(sample_df, states_df,
                    on = 'State',
                    how = 'left')
sample_df
```

# I rename column state to state_abbreviation

```python
sample_df = sample_df.rename(columns = {'State': 'State Abbreviation'})
sample_df.head(3)
```

# I place the state_name column after the state_abbreviation column

```python
# I copy and save the state_name column, then remove it from its original position

state_name_col = sample_df.pop('State Name')

# I insert the column again in its new position
    # The first argument (3) specifies the new position for the column
    # The second argument ('State Name') specifies the column name
    # The third argument (col_a) contains the column’s data

sample_df.insert(3, 'State Name', state_name_col)
sample_df.head(4)
```
